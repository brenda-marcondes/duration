```python
#### Duration   

from datetime import date
from datetime import datetime
import pandas as pd

###### INPUT

valor_nominal = 1000  
cupom =  0.05  
anos = 3
lista_anos = list(range(1,anos+1))
juros = 0.06


#Fluxo Mensal

fluxo_de_caixa = []

fx_anos = anos

while fx_anos > 1:
    fluxo_ano = valor_nominal * cupom
    
    fluxo_de_caixa.append(fluxo_ano)
    
    fx_anos = fx_anos - 1
else:
    fluxo_ano = valor_nominal + (valor_nominal * cupom)
    fluxo_de_caixa.append(fluxo_ano)
    fx_anos = fx_anos - 1
    
print("Fluxo mensal: ", fluxo_de_caixa)


###Fator de Desconto

fator_de_desconto = []

anos_ftdesconto = anos
i = 1

while anos_ftdesconto >= 0:
    fator_de_desconto_mes = 1 / (1+juros)**i
    i = i+1
    
    fator_de_desconto.append(fator_de_desconto_mes)
    anos_ftdesconto = anos_ftdesconto - 1

print("Fator de desconto: ", fator_de_desconto)


### VALOR PRESENTE
valor_presente = [fc * fd for fc, fd in zip(fluxo_de_caixa, fator_de_desconto)]
print("Valor presente anual: ", valor_presente)
valor_presente_soma = sum(valor_presente)
print("Valor Presente Total", valor_presente_soma)


### Peso VP

proporcao_valor_presente = [vp / valor_presente_soma for vp in valor_presente]
print(proporcao_valor_presente)

### PesoVP x Tempo

PPV_Tempo = [pvpresente * list_anos for pvpresente, list_anos in zip(proporcao_valor_presente,lista_anos)]
print(PPV_Tempo)

duration_Macaulay = sum(PPV_Tempo)
print("Duration Macaulay: ",duration_Macaulay)

duration_modificada = duration_Macaulay / (1+juros)
print("Duration Modificada:", duration_modificada, "Retorno em anos")




### Criando o DataFrame
df = pd.DataFrame({
    'Ano': lista_anos,
    'Fluxo de Caixa': fluxo_de_caixa,
    'Fator de Desconto': fator_de_desconto[:len(fluxo_de_caixa)],
    'Valor Presente': valor_presente,
    'Proporção VP': proporcao_valor_presente,
    'Peso VP x Tempo': PPV_Tempo
})

df.head(30)
df.to_excel("fluxo_de_caixa.xlsx", index=False)
```
